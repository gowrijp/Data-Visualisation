```python
#Pyechart is a python echarts plotting library with more than 30 kinds of charts
from pyecharts import options as opts
from pyecharts.charts import Liquid
from pyecharts.globals import SymbolType
```

```python
#Generate Water Polo chart
c = (
    Liquid()
    .add("Completion", [0.6])
    .set_global_opts(title_opts=opts.TitleOpts(title="% of sales target achieved",pos_left="center"))
)
c.render_notebook()
```

```python
#Generate file in html format
#Render the water polo map as html by the render() method
c = (
 Liquid()
 .add("Completion", [0.6])
 .set_global_opts(title_opts=opts.TitleOpts(title="% of sales target achieved",pos_left="center"))
 .render("roundwaterpolo.html")
)
```

```python
#Removing outline of the chart
c = (
    Liquid()
    .add("Completion", [0.6], is_outline_show=False)
    .set_global_opts(title_opts=opts.TitleOpts(title="% of sales target achieved",pos_left="center"))
)
c.render_notebook()
```

```python
#Customize the color of the water polo
c = (
    Liquid()
    .add("Completion", [0.6], is_outline_show=False, color=['#7733FF'])
    .set_global_opts(title_opts=opts.TitleOpts(title="% of sales target achieved",pos_left="center"))
)
c.render_notebook()
```

```python
#Customize the shape of the water polo chart — round,rectangular, arrow,diamond,triangle
c = (
    Liquid()
    .add("Completion", [0.6], is_outline_show=False, color=['#7733FF'],shape=SymbolType.ROUND_RECT)
    .set_global_opts(title_opts=opts.TitleOpts(title="% of sales target achieved",pos_left="center"))
)
c.render_notebook()
```
